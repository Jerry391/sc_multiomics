```python
# clustering the RNA
n = 2194
rna_pc = adata[3 * n : 4 * n, adata.var.modality == "Gene Expression"].copy()
rna_pc.var
# sc.tl.pca(rna_pc)
sc.pp.neighbors(rna_pc, n_pcs=40)
sc.tl.umap(rna_pc)
sc.tl.pca(rna_pc, svd_solver='arpack')

# rna_pc.obsm['X_pca'].shape

new_adata = sc.AnnData(X=rna_pc.obsm['X_pca'])

new_adata.obs = rna_pc.obs.copy()
new_adata.var["modality"] = "RNA_PC"
rna_pc = new_adata
rna_pc
```

```python
# clustering the ATAC
n = 2194
atac_pc = adata[4 * n :, adata.var.modality == "Gene Expression"].copy()
atac_pc.var
# sc.tl.pca(rna_pc)
# sc.pp.neighbors(atac_pc, n_pcs=40)
sc.tl.umap(atac_pc)
sc.tl.pca(atac_pc, svd_solver='arpack')

# rna_pc.obsm['X_pca'].shape

new_adata = sc.AnnData(X=atac_pc.obsm['X_pca'])

# 如果你希望保留原始AnnData对象的观测信息和变量信息，你可以像这样复制它们到新对象中
new_adata.obs = atac_pc.obs.copy()
new_adata.var["modality"] = "RNA_PC"
atac_pc = new_adata
```

```python
# k-means
# use k-means to cluster the adata
from sklearn.cluster import KMeans

def kmeans_cluster(adata, n_clusters=10):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(adata)
    adata.obs["kmeans"] = kmeans.labels_.astype(str)
    return adata

n = 2194
atac_pc = adata[4 * n :, adata.var.modality == "Gene Expression"].copy()
atac_pc.var
# sc.tl.pca(rna_pc)
sc.pp.neighbors(atac_pc, n_pcs=40)
sc.tl.umap(atac_pc)
sc.tl.pca(atac_pc, svd_solver='arpack')

# kmeans_cluster(rna_pc, n_clusters=10)
kmeans_cluster(atac_pc, n_clusters=10)
```
